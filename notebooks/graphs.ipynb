```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x = np.linspace(-1, 1, 500)

# Calculate the tanh values
y = np.tanh(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='tanh(x)', color='blue')
plt.title('Graph of tanh(x) from -1 to 1')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the GELU function
# def gelu(x):
#     return 0.5 * x * (1 + scipy.special.erf(x / np.sqrt(2)))

# Define the GELU function without using scipy
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

# Define the x range
x = np.linspace(-1, 1, 500)

# Calculate the GELU values
y = gelu(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='GELU(x)', color='green')
plt.title('Graph of GELU(x) from -1 to 1')
plt.xlabel('x')
plt.ylabel('GELU(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def relu(x):
    return np.maximum(0, x)

def relu6(x):
    return np.minimum(np.maximum(0, x), 6)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def selu(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * np.where(x > 0, x, alpha * (np.exp(x) - 1))

def softplus(x):
    return np.log(1 + np.exp(x))

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

# Define the x range
x = np.linspace(-3, 3, 500)

# Calculate the activation values
y_relu = relu(x)
y_relu6 = relu6(x)
y_elu = elu(x)
y_selu = selu(x)
y_softplus = softplus(x)
y_gelu = gelu(x)

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(x, y_relu, label='ReLU', color='blue')
plt.plot(x, y_relu6, label='ReLU6', color='orange')
plt.plot(x, y_elu, label='ELU', color='green')
plt.plot(x, y_selu, label='SELU', color='red')
plt.plot(x, y_softplus, label='Softplus', color='purple')
plt.plot(x, y_gelu, label='GELU', color='cyan')
plt.title('Activation Functions')
plt.xlabel('x')
plt.ylabel('Activation')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

```
